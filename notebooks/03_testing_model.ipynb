```python
# Cargar el modelo entrenado
model = YOLO("C:/Users/marck/OneDrive/Documentos/Proyecto_ia/runs/detect/train/weights/best.pt")

# Inicializar la cámara (0 es la webcam por defecto)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("❌ No se pudo abrir la cámara")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("❌ No se pudo leer el frame")
        break

    # Realizar predicción
    results = model.predict(source=frame, show=False, conf=0.7)

    # Dibujar resultados en el frame
    annotated_frame = results[0].plot()

    # Mostrar el resultado
    cv2.imshow("YOLOv11 Detección en Tiempo Real", annotated_frame)

    # Salir con 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
